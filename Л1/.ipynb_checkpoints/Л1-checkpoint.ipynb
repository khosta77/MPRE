## Лабараторная номер 1

### Задача:
В момент времени t = 0 волновая функция, описывающая квантовое состояние микрочастицы, движущейся вдоль оси х имеет вид: 

$\psi(x,0) = A exp(-\frac{x^2}{a^2}) + ibx)$

Здесь: A,a,b - известные действительные константы. **Определить** зависимости от координаты х:

- действительной части волновой функции;
- квадрата модуля волновой функции.

**Построить** эти зависимости.
### Теория:
Формула Эйлера связывает комплексную экспоненту с тригонометрическими функциями. Она утверждает, что для любого вещественного числа x выполнено следующее равенство:

$\psi = cos(x) + i*sin(x).$

Тогда решение:

$y = A*exp(\frac{-x^2}{a^2})*(cos(b*x) + i*sin(b*x))$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = 5
a = 2.2
b = 4
x = np.linspace(-5, 5, 100)
i = 1j
```

```python
# Вычисляем волновую функцию
Psis = A*np.exp((-x**2)/(a**2))*(np.cos(b*x) + i*np.sin(b*x))

# Извлекаем реальную часть
PsisReal = Psis.real

# Берем квадрат по модулю
PsisAbs = np.abs(Psis**2)
```

**Ошибку выводит в следствии того, что мы в начале пытаемся построить график с мнимыми величинами**

```python
plt.figure(figsize=(15,10))
plt.plot(x, PsisAbs, "g", label='|Psis|^2')
plt.plot(x, Psis, "r--", linewidth = 4, label='Psis')
plt.plot(x, PsisReal, "b", label='real(Psis)')
ax = plt.gca()
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 25])
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Psis")
plt.grid()
plt.legend()
plt.show()
```
