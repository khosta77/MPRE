$\psi(r, t) = Aexp(\frac{-r}{a})*exp(-i*(\frac{E}{h})*t)$

### Здесь:

r - расстояние силового центра;
a - известная константа( a = 0.529177*10^(-10));
E - полная энергия частицы независящая от времени.
1. Определить занечение постоянной А.
2. Построить графики (действительной, мнимой частей и модуля) нормиравнной сферической
гармоники ВФ для s- и p- квантовых состояний.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np
import math
```

```python
def sph2cart(theta, phi, r):
    x = r * np.cos(phi) * np.cos(theta)
    y = r * np.cos(phi) * np.sin(theta)
    z = r * np.sin(phi)
    return x, y, z
```

```python
def plot(Azimuth, Elevation, r):
    x, y, z = sph2cart( Azimuth, Elevation, r)
    fig = plt.figure()
    ax = fig.add_subplot( 111, projection='3d')
    ax.plot_surface(x, y, z, linewidth = 0.5, edgecolors = 'k')
```

```python
i = 1j
phi = np.linspace(0, 2 * math.pi)
theta = np.linspace(0, math.pi)
Phi, Theta = np.meshgrid(phi, theta)
```

```python
Azimuth = Phi
Elevation = (math.pi/2) - Theta;
r = np.sqrt(1/(4 * math.pi));
rPm0 = (1/2) * np.sqrt(3/math.pi) * np.cos(Theta);
rPm1 = (-1/2)* np.sqrt(3/(2 * math.pi)) * np.sin(Theta) * np.exp(i * Phi);
```

```python
plot(Azimuth, Elevation, r)
```

```python
plot(Azimuth, Elevation, np.abs(rPm0))
```

```python
plot(Azimuth, Elevation, np.abs(rPm1))
```

```python
plot(Azimuth, Elevation, np.abs(rPm1.imag))
```

```python
plot(Azimuth, Elevation, np.abs(rPm1.real))
```
